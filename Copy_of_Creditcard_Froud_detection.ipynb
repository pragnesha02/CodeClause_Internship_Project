Data Visualization

```python
plt.scatter(data['Time'],data['Amount'],color='g')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.title('scatter plot')
print(plt.show())
```

```python
plt.plot(data['Class'],data['V20'],color='y')
plt.xlabel('Class')
plt.ylabel('V20')
plt.title('Line plot')
print(plt.show())
```

```python
plt.hist(data['V15'],color='silver')
plt.title('Histogram')
print(plt.show())
```

```python
plt.bar(data['Time'],data['V11'],color='b')
plt.xlabel('Time')
plt.ylabel('V11')
plt.title('Bar plot')
print(plt.show())
```
